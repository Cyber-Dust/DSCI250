```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
titanic
```

```python
# comparison bar chart, compares an x and y axis
comparisonPlot = sns.catplot(x= 'sex', y ='survived',data=titanic,kind='bar')
```

```python
# RELATIONSHIPS (done with quantitative values)
# scatterplot is useful
# https://seaborn.pydata.org/generated/seaborn.scatterplot.html
diamonds = pd.read_csv("diamonds.csv")
diamonds
```

```python
# the hue dimension will control the color of data points
sns.scatterplot(x="carat", y="price", hue="clarity",data=diamonds)
```

```python
planets = pd.read_csv("planets.csv")
planets
```

```python
ax = sns.scatterplot(x="distance", y="orbital_period",sizes=(10,200),size="mass",data=planets,hue="year")
# smallest mass will have a value of 10, and largest will have a value of 300
```
